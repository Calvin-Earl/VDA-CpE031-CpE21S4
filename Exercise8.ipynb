```python
from google.colab import files
import pandas as pd
uploaded = files.upload()

data = pd.read_csv('most_subscribed_youtube_channels.csv')
data.head(50)
```

```python
data.to_csv('most_subscribed_youtube_channels.csv')
youtube = pd.DataFrame(data)
ytchannel = youtube.head(50)
ytchannel
```

```python
#1. Box plot

plt.boxplot(ytchannel['started'])
plt.title("Year started boxplot")
plt.ylabel("year")
plt.show()
```

```python
#2. Pie chart

category_counts = ytchannel.groupby('category')['category'].count()
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Youtube channels Video Categories')
plt.show()
```

```python
#3. Bar graph

channels = ytchannel['Youtuber'][:4]
subcount = ytchannel['subscribers'][:4]

plt.bar(channels, subcount)
plt.title("Most Subscribed Youtube Channels")
plt.xlabel("Youtube Channel")
plt.ylabel("Subscribers")
plt.xticks(rotation=45, ha='right')

plt.show()
```
