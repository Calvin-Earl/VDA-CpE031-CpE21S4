```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from google.colab import drive
drive.mount('/content/drive')

path = '/content/drive/MyDrive/datset.csv'
data = pd.read_csv('/content/drive/MyDrive/datset.csv')

data
```

```python
slope, intercept = np.polyfit(data['x'][:-1],data['y'][:-1],1)
y_pred = slope * data['x'][:-1] + intercept

plt.scatter(data['x'][:-1],data['y'][:-1])
plt.plot(data['x'][:-1],y_pred,color='red')
plt.show()
```

Upon plotting the given dataset, we can observe that the dots are close to each other. However, we can also observe that the line seems to be inverted, as it starts at the top left going down right. And after solving for r using the Pearson correlation coefficient formula, I got a value of -0.95, which classifies as very strong negative correlation. We can therefore conclude that the correlation between the x variable and the y variable has a very strong negative correlation with an r coefficient of -0.95.
