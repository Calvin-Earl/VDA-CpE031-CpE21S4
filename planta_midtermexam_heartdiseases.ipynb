CSV 1 - heart_2022_no_nans

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files

uploaded = files.upload()
df = pd.read_csv('heart_2022_no_nans.csv')

df.head(50)
```

```python
heart = pd.DataFrame(df.head(50))
heart
```

Question 1: What is the most appropriate visualization method to compare the the percentage of heart attacks for both genders?

```python
malepercent = heart.groupby('Sex')['HadHeartAttack'].count()
plt.pie(malepercent, labels=malepercent.index, autopct='%1.1f%%')
plt.title('Male vs Female Heart attacks')
plt.show()
```

Observarion: In order to display the percentages of heart attacks for both genders, I used the pie chart visualization to represent the data. Additionally, I used autopct = '&1.1f& to show the percentage of each slice in one decimal only. According to the data provided by the given dataset, females had more heart attacks compared to males, with 30 (60%) surveyed females having heart attacks while only 20 (40%) surveyed males had heart attacks.  

Question 2: What would be the most appropriate graph to use to compare the average sleep hours of both male and female

```python
average_sleep = heart.groupby('Sex')['SleepHours'].mean()
plt.bar(average_sleep.index, average_sleep.values)

plt.xlabel("Sex")
plt.ylabel("Average Sleep Hours")
plt.title("Comparison of Average Sleep Hours between Male and Female")
plt.xticks(average_sleep.index, ['Female', 'Male'])

plt.show()
```

Observation: I used the bar graph to compare the average sleep hours of both male and female, which I was able to do by getting the average sleep hours of both genders, and plotting it to the graph. According to the data, men get more hours of sleep compared to women, with men having an average of 8 hours of sleep while women only having an average of 7 hours of sleep.

CSV 2 - heart_2022_with_nans

```python
uploaded = files.upload()
df = pd.read_csv('heart_2022_with_nans.csv')
```

```python
heart2 = pd.DataFrame(df.head(50))
heart2
```

Question 1: What data vizualization would be the most appropriate for displaying the mean BMI of males?

```python
import seaborn as sns

male_data = heart2[heart2['Sex'] == 'Male']

plt.figure(figsize=(8, 6))
sns.boxplot(y=male_data['BMI'].dropna())

plt.title('Distribution of BMI for Males')
plt.ylabel('BMI')
plt.show()
```

Observation: I used the boxplot graph to display the average BMI of males. In the BMI column, some of the data is displayed 'NaN', and in order to exclude them, I used .dropna(). According to the graph, the average BMI of males ranges from 18 to 34, and the median is 26.5

Question 2: What would be the most appropriate data visualization to display the average weight of females?

```python
female_data = heart2[heart2['Sex'] == 'Female']

plt.figure(figsize=(8, 6))
sns.boxplot(y=female_data['WeightInKilograms'].dropna())

plt.title('Distribution of Weight for Females')
plt.ylabel('Weight')
plt.show()
```

Observation: I used yet another boxplot to display the weight of females. In the Weight column, some of the data is displayed 'NaN', and in order to exclude them, I used .dropna(). According to the graph, the weight in kilograms of females ranges from 38 to 110, and the median is 65.
