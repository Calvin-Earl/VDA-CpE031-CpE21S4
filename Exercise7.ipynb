```python
#3. Creating a dataframe from a dictionary of lists
data = {'rank' : [1, 2, 3, 4, 5], 'country_rank' : [1, 2, 3, 4, 5],
        'player_name' : ['Cookiezi', 'Angelsim', 'Rafis', 'Vaxei', 'hvick225'],
        'country' : ['South Korea', 'South Korea', 'Poland', 'United States', 'Taiwan'],
        'performance_points' : [13450, 12950, 12942, 12735, 12421]}
ranking = pd.DataFrame(data)
print(ranking)
```

```python
#4. Creating a dataframe from a list of dictionaries
data = [
    {'rank': 1, 'country_rank': 1, 'player_name': 'Cookiezi', 'country': 'South Korea', 'performance_points': 13450},
    {'rank': 2, 'country_rank': 2, 'player_name': 'Angelsim', 'country': 'South Korea', 'performance_points': 12950},
    {'rank': 3, 'country_rank': 3, 'player_name': 'Rafis', 'country': 'Poland', 'performance_points': 12942},
    {'rank': 4, 'country_rank': 4, 'player_name': 'Vaxei', 'country': 'United States', 'performance_points': 12735},
    {'rank': 5, 'country_rank': 5, 'player_name': 'hvick225', 'country': 'Taiwan', 'performance_points': 12421}
]
ranking = pd.DataFrame(data)
print(ranking)
```

```python
#5. Selecting dataframe columns
print(ranking['player_name']) # Single

print()

print(ranking[['player_name', 'country_rank']]) # Multiple
```

```python
#6. Selecting dataframe rows
print(ranking.loc[2]) # Single

print()

print(ranking.loc[[2, 4]]) # Multiple
```
