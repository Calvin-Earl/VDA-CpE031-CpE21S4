```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from google.colab import files

uploaded = files.upload()
data = pd.read_csv('Technological-Products-Sample-Data - Tech Products.csv')
data
```

```python
tech = pd.DataFrame(data)
tech
```

```python
#1. plotting the dataset using Line graph

xlabel = tech['Date of Release']
ylabel = tech['Price (USD)']
plt.figure(figsize=(10, 6))
plt.plot(xlabel, ylabel)
plt.xlabel('Date of Release')
plt.ylabel('Price')
plt.xticks(rotation=45, ha='right')
plt.title("Price of Technological Products")
plt.show()
```

```python
#2. Plotting the total sales using line plot (with custom properties)

xlabel = tech['Date of Release']
ylabel = tech['Price (USD)']
plt.figure(figsize=(10, 6))
plt.plot(xlabel, ylabel, linestyle = 'dotted', color = 'red',
        marker = 'o', mec = 'red', linewidth = 3, label = ['Sales'])
plt.xlabel('Year')
plt.ylabel('Sales')
plt.xticks(rotation=45, ha='right')
plt.title("Total sales of Technological Products")
plt.legend(loc = 'lower right')
plt.show()
```

```python
#3. Plotting the product sales using multiline plot
products = tech['Device'].unique()

plt.figure(figsize=(10, 6))

for product in products:
    product_data = tech[tech['Device'] == product]
    plt.scatter(product_data['Date of Release'], product_data['Price (USD)'], label=product)

plt.xlabel('Date of Release')
plt.ylabel('Price (USD)')
plt.title('Product Sales Over Time')
plt.xticks(rotation=45, ha='right')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

```python
#4. Plotting the brand sales using bar graph

xlabel = tech['Brand']
ylabel = tech['Price (USD)']
plt.figure(figsize=(10, 6))
plt.bar(xlabel, ylabel)
plt.xlabel('Brand')
plt.ylabel('Sales')
plt.xticks(rotation=45, ha='right')
plt.title("Sales of each Technological Products")
plt.show()
```

```python
#5. Using Histogram to see the most common Country of Origin
plt.hist(tech['Country of Origin'])
plt.xlabel('Country of Origin')
plt.ylabel('Frequency')
plt.title("Most common Country of Origin")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#6. Using pie chart to show each Brand's Country of Origin

countrynum = tech.groupby('Brand')['Country of Origin'].nunique()
plt.pie(countrynum, labels=countrynum.index, autopct='%1.1f%%')
plt.title("Country of Origin of each Brand")
plt.show()
```
